```python
import torch
import random

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
import torch.nn.functional as F
```

```python
words = open('names.txt','r').read().splitlines()
```

```python
chars = sorted((set(''.join(words))))
stoi = {c:i+1 for i,c in enumerate(chars)}
stoi['.'] = 0
stoi
```

```python
xs, ys = [], []

for word in words:
    new_word = ['.'] + list(word) + ['.']

    for w1, w2 in zip(new_word, new_word[1:]):
        biagram = (w1, w2)
        xs.append(stoi[w1])
        ys.append(stoi[w2])
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()

# Weight Initialization
g = torch.Generator().manual_seed(2147483647)
w = torch.randn(27, 27, generator=g, requires_grad=True) # 27x27
```

Forwad Pass

```python
for k in range(1000):
    # forward pass
    xenc = F.one_hot(xs, num_classes=27).float()
    logits = xenc @ w # log-counts
    counts = logits.exp() # counts
    probs = counts / counts.sum(1, keepdim=True) # probabilities

    # loss calculation
    loss = -probs[torch.arange(num), ys].log().mean() + 0.01 * (w**2).mean() # L2 regularization
    print(loss)

    # Backward pass
    w.grad = None
    loss.backward()

    # update Weights
    w.data -= 0.2 * w.grad
```
